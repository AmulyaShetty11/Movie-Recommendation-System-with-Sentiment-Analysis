## Extracting features of 2022 movies from Wikipedia

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2022"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
# this two rows are going to coz further error coz tmbd doesn't have movie having this title
# you can also auttomate this process by modifying the function if present in huge number
df2['Title'][4] = 'Apollo 10½: A Space Age Childhood'
```

```python
from tmdbv3api import TMDb
import requests
tmdb = TMDb()
tmdb.api_key = '61eba94e833683f5a5dad5a53c3c9c4a' #your api key will be here
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
    data_json = response.json()
    if data_json['genres']:
        genre_str = " " 
        for i in range(0,len(data_json['genres'])):
            genres.append(data_json['genres'][i]['name'])
        return genre_str.join(genres)
    else:
        np.NaN
```

```python
df['genres'] = df['Title'].map(lambda x: get_genre(str(x)))
```

```python
df.head()
```

```python
df_2022 = df[['Title','Cast and crew','genres']]
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2022['director_name'] = df_2022['Cast and crew'].map(lambda x: get_director(str(x)))
```

```python
df_2022.head()
```

```python
def get_actor1(x):
    return (x.split('screenplay); ')[-1].split(', ')[0])
```

```python
df_2022['actor_1_name'] = df_2022['Cast and crew'].map(lambda x : get_actor1(str(x)))
```

```python
def get_actor2(x):
    if len((x.split('screenplay); ')[-1]).split(', ')) < 2:
        return np.NAN
    else :
        return (x.split('screenplay); ')[-1].split(', ')[1])
```

```python
df_2022['actor_2_name'] = df_2022['Cast and crew'].map(lambda x: get_actor2(str(x)))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2022['actor_3_name'] = df_2022['Cast and crew'].map(lambda x: get_actor3(str(x)))
```

```python
df_2022.head()
```

```python
df_2022 = df_2022.rename(columns={'Title':'movie_title'})
```

```python
new_df22 = df_2022.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
new_df22.isna().sum()
```

```python
new_df22['actor_2_name'] = new_df22['actor_2_name'].replace(np.nan, 'unknown')
new_df22['actor_3_name'] = new_df22['actor_3_name'].replace(np.nan, 'unknown')
new_df22.dropna(inplace=True)
```

```python
new_df22.isna().sum()
```

```python
new_df22['movie_title'] = new_df22['movie_title'].str.lower()
```

```python
new_df22['comb'] = new_df22['actor_1_name'] + ' ' + new_df22['actor_2_name'] + ' '+ new_df22['actor_3_name'] + ' '+ new_df22['director_name'] +' ' + new_df22['genres']
```
